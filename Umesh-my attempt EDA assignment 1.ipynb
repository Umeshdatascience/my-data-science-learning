```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("student-mat.csv")
```

```python
df
```

```python
#summary statistics
df.describe()
```

```python
#the complete information about the data set
df.info()
```

```python
#inspect missing values in the dataset
print(df.isnull().values.sum())

#Replace the ' 's with NaN
df = df.replace(" ",np.NaN)

#Impute the missing values with mean imputation
df = df.fillna(df.mean())

#Count the number of NaNs in the data set to verify
print(df.isnull().values.sum())
```

```python
# seaborn histogram
sns.distplot(df['age'], hist=True, kde=False,
             bins=12, color= 'blue',
             hist_kws={'edgecolor': 'black'})
#Add labels
plt.title('Age count')
plt.xlabel('Age')

plt.ylabel('Count')


```

```python
#seaborn Scatterplot
sns.scatterplot(x=df['absences'], y=df['age'])
```

```python
#Line plot
sns.lineplot(x='absences', y='age', data=df)
```

```python
#box plot
sns.boxplot(y= 'age', data= df, x= 'sex')
plt.xlabel('SEX')
```

```python
#this is heat map pearson correlation matrix
corrmat = df. corr()
f, ax = plt.subplots(figsize=(16, 12))
sns.heatmap(corrmat, vmax=.8, square=True);
```

```python
plt.figure(figsize=(20,20))
plt.title('Pearson Correlation of Features', size=15)
colormap = sns.diverging_palette(10, 220, as_cmap = True)
sns.heatmap(df.corr(),
            cmap = colormap,
            square = True, 
            annot = True,
            linewidths=0.1, vmax=1.0, linecolor='white',
            annot_kws={'fontsize':12})

plt.show()
```
